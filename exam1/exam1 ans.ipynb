```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from openpyxl import Workbook, load_workbook 
from openpyxl.utils import get_column_letter
```

```python
wb = load_workbook('exam excel.xlsx')
ws = wb.active
ws = wb[wb.sheetnames[6]]
```

```python
x = [ws['B9'].value, ws['B11'].value, ws['B13'].value, ws['B15'].value, ws['B17'].value, ws['B19'].value, ws['B21'].value,]
y = [ws['D9'].value, ws['D11'].value, ws['D13'].value, ws['D15'].value, ws['D17'].value, ws['D19'].value, ws['D21'].value,]
y2 = [ws['D10'].value, ws['D12'].value, ws['D14'].value, ws['D16'].value, ws['D18'].value, ws['D20'].value, ws['D22'].value, ]
plt.plot(x, y, marker = 'o', label = 'Male')
plt.plot(x, y2, marker = 'o', color = 'orange', label = 'Female')
plt.legend()
plt.title('Marriage Registration')


```
