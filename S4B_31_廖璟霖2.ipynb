```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1*np.pi, 4*np.pi, 100)

title = ['$f(x) = A \sin ( \omega x + \phi)$' , 
         '$f(x) = A \cos ( \omega x + \phi)$',
         '$f(x) = A \sin ( \omega x + \phi)$ + $f(x) = A \cos ( \omega x + \phi)$'
        ]

def a():
    plt.xticks([-np.pi, 0,  np.pi,  2*np.pi, 3*np.pi ,4*np.pi],
          ['$-\pi$', '$0$', '$\pi$', '$2\pi$', '$3\pi$' ,'$4\pi$'])
    plt.yticks([-np.pi, 0, np.pi],
          ['$-\pi$', '$0$', '$\pi$'])
    
    plt.xlabel('X', color = 'black', size = '20')
    plt.ylabel('Y', color = 'black', size = '20')
    
    plt.grid(color='g', linestyle='--', alpha = 0.5, linewidth=1)

def plotsine(A = 1, omega = 1, phi = 0):
    a()
    y = A * np.sin(omega * x + phi)
    plt.plot(x, y, label = '$\sin$', color = 'blue')
    plt.title( title[0], ha = 'center', color = 'blue', size='15')
    plt.legend()
    plt.tight_layout()
    plt.fill_between(x ,y, alpha = 0.1)
        
def plotcosine(A = 1, omega = 1, phi = 0):
    a()
    y = A * np.cos(omega * x + phi)
    plt.plot(x, y, label = '$\cos$', color = 'red')
    plt.title( title[1], ha = 'center', color = 'red', size = '15')
    plt.legend()
    plt.tight_layout()
    
def plotboth(A = 1, omega = 1, phi = 0):
    a()
    y = A * np.sin(omega * x + phi)
    y_cos = A * np.cos(omega * x + phi)
    plt.plot(x, y, label = '$\sin$', color = 'blue')
    plt.plot(x, y_cos, label = '$\cos$', color = 'red')
    #plt.title( title[2], color = 'purple', size = '15')
    plt.legend()
```

```python
plotsine()
```

```python
plotcosine()
```

```python
plt.subplot(221)
plotsine()
plt.subplot(222)
plotcosine()
```

```python
plotboth()
```
