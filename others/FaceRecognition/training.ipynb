```python
import cv2
import numpy as np
import os
```

```python
path = 'images/p0'
NumbersofImage = 0
for file in os.listdir(path):
    fullname = os.path.join(path, file)
    if os.path.isdir(fullname):
        print(f"{file}<DIR>")
    elif file.endswith(".jpeg"):
        print(f'{file}')
        NumbersofImage += 1
```

```python
NumbersofImage
```

```python
images = []
labels = []   #  append(0)   means the people in images is 0

for index in range(NumbersofImage):
    filename = f'images/p0/{index}.jpeg'
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # if index < 3:
    #     cv2.imshow('check', img)
    #     cv2.waitKey(0)
    #     cv2.destroyWindow('check')
    #     cv2.waitKey(1)
    images.append(img)
    labels.append(0)
```

```python
print('training......')
model = cv2.face.LBPHFaceRecognizer_create()
model.train(np.asarray(images), np.asarray(labels))
model.save('p0.data')
print('training done')
```

```python
import cv2
model = cv2.face.LBPHFaceRecognizer_create()
model.read('p0.data')
print('load training data done')



ESC = 27 #ASCII code of ESC is 27
cap = cv2.VideoCapture(0)#0 is the first camera of your computer
cv2.namedWindow('test', cv2.WINDOW_NORMAL)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt2.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
names = ['kenny']  # name of the face

while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.4, 3)
    
    for (x, y, w, h) in faces:
        
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)
        face_image = cv2.resize(gray[y: y+h, x: x+w], (400,400))
       
        try:
            val = model.predict(face_image)
            print(f'label:{ val[0]}, confidence:{ val[1]}')
            if val[1] < 50:
                cv2.putText(frame, names[val[0]], (x, y - 10), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 3 )
        except:
            pass
            
    cv2.imshow('test', frame)
      
    if cv2.waitKey(10) == ESC:
        break
        
cv2.destroyAllWindows()
cv2.waitKey(1)   
cap.release()
```
